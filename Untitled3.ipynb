```python
mask = np.random.rand(len(df)) < 0.8
train = cdf[mask]
test = cdf[~mask]
print(train)
print(test)
```

```python
fig = plt.figure()
ax1=fig.add_subplot(111)
ax1.scatter(train.ENGINESIZE, train.CO2EMISSIONS , color='red')
ax1.scatter(test.ENGINESIZE, test.CO2EMISSIONS , color='yellow')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
train_x= np.asanyarray(train[['ENGINESIZE']])
train_x= np.asanyarray(train[['CO2EMISSIONS']])
reg.fit(train_x , train_x)
print("coeffisiont",reg.coef_)
print("intercept", reg.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS , color='blue')
plt.plot(train_x , (reg.coef_[0][0]*train_x +  reg.intercept_[0]),'-r')
```
